```python
## script to recover parameters for all 3 models for the new dimension learning task ## 

# get packages
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# get scripts
import models.FRL as FRL
import models.CAFRL as CAFRL
import models.SAFRL as SAFRL
import models.CAFRL_A as CAFRL_A
import models.SAFRL_A as SAFRL_A
import scripts.modelling as modelling

# initialise lists and variables
models = [CAFRL, SAFRL, CAFRL_A, SAFRL_A]
param_names = ['alpha', 'beta', 'theta0', 'epsilon']
param_names_2 = ['alpha', 'beta', 'gamma']
param_names_3 = ['alpha', 'beta', 'epsilon', 'gamma']
transforms = ['sigmoid', 'exp', None, 'sigmoid']
transforms_2 = ['sigmoid', 'exp', 'sigmoid', 'sigmoid']
nP_model = [3, 4, 3, 4]

# how many participants do you want to sim
n = 500

# load fits to simulate from
with open('data/modelling/fits/frl/all_!2.pkl', 'rb') as f:
    frl = pickle.load(f)
with open('data/modelling/fits/cafrl/all_!2.pkl', 'rb') as f:
    cafrl = pickle.load(f)
with open('data/modelling/fits/safrl/all_!2.pkl', 'rb') as f:
    safrl = pickle.load(f)
with open('data/modelling/fits/cafrl_A/all_!2.pkl', 'rb') as f:
    cafrl_A = pickle.load(f)
with open('data/modelling/fits/safrl_A/all_!2.pkl', 'rb') as f:
    safrl_A = pickle.load(f)

# lists of means and covariance matrices
frl_u, cafrl_u, safrl_u, cafrl_A_u, safrl_A_u =  frl['u'], cafrl['u'], safrl['u'], cafrl_A['u'], safrl_A['u']
frl_v2, cafrl_v2, safrl_v2, cafrl_A_v2, safrl_A_v2 =  frl['v2'], cafrl['v2'], safrl['v2'], cafrl_A['v2'], safrl_A['v2']
m_list = [cafrl_u, safrl_u,  cafrl_A_u, safrl_A_u]
v2_list = [cafrl_v2, safrl_v2, cafrl_A_v2, safrl_A_v2]
```

```python
# now lets recover parameters for each model

# first loop over each model
for model in models:

    # get the index of the model in our models list
    i = models.index(model)

    ## generate params from multivariate normal
    # get appropraite means and covariance matrices from lists
    m = m_list[i]
    v2 = v2_list[i]

    # set distribtuion
    dist = (m,v2)
    
    # generate params from dist
    parameters = np.random.multivariate_normal(dist[0], dist[1], n)
    nP = parameters.shape[1]

    # transform parameters 
    # first get the right transforms for the mdoel
    if model in [CAFRL_A, SAFRL_A]:
        t = transforms_2
    else:
        t = transforms
    
    for x in range(nP_model[i]):
        if t[x] == 'sigmoid':
            parameters[:,x] = 1/(1+np.exp(-parameters[:,x]))
        elif t[x] == 'exp':
            parameters[:,x] = np.exp(parameters[:,x])

    # run simulation
    sim = modelling.simulate(model = model.simulate, model_args = 0, rng = None, seed = 999, params=parameters, dist=None, N=n, transforms = None, subjects = None)

    # now fit the simmed data to the model
    fit = modelling.fit(data = sim['data'], nP = nP_model[i], fit_func= model.fit, fit_args = ['reward', 'choice', 'stim_hist', 'dimension1'], transforms = t[0:nP_model[i]], EM_iter= 100)

    # now plot and calc correlations

    # plotting settings
    sns.set_theme(rc={'figure.figsize':(8,25)})
    sns.set_style(style='white')

    # build figure space
    fig, axs = plt.subplots(4)

    # plot titles, capitalising parameter names for the params in the model
    if model in [CAFRL_A]:
        param_name = param_names_2
    elif model in [SAFRL_A]:
        param_name = param_names_3
    else:
        param_name = param_names
    plot_titles = [param.title() for param in param_name][0:nP_model[i]]

    # loop over the number of parameters and plot recoverd against ground truth
    for j in range(nP_model[i]):

        # get correlation and "corr = x" into the title
        correlation = np.round(np.corrcoef(fit['m'][:,j],parameters[:,j])[0,1], decimals = 2)
        plot_titles[j] = plot_titles[j] + ", corr = " + str(correlation)

        sns.scatterplot(x=fit['m'][:,j], y=parameters[:,j], ax=axs[j])
        axs[j].set_xlabel("Fitted Parameters")
        axs[j].set_ylabel("Simulated Parameters")
        axs[j].set_title(plot_titles[j])

    plot_name = 'model_' + str(i+1) + '_all_!2.png'
    fig.savefig(plot_name)
```
