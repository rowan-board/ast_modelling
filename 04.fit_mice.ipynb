```python
import numpy as np 
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import pickle

import scripts.modelling as modelling
import models.CAFRL as CAFRL
import models.SAFRL as SAFRL
import models.SAFRL_A as SAFRL_A
import models.CAFRL_A as CAFRL_A
import models.FRL as FRL
```

```python
# load model data
with open('data/modelling/datalists/ghc_r0.pkl', 'rb') as f:
    data_ghc = pickle.load(f)

with open('data/modelling/datalists/ghd_r0.pkl', 'rb') as f:
    data_ghd = pickle.load(f)

with open('data/modelling/datalists/isoc_r0.pkl', 'rb') as f:
    data_isoc = pickle.load(f)

with open('data/modelling/datalists/isod_r0.pkl', 'rb') as f:
    data_isod = pickle.load(f)

with open('data/modelling/datalists/all_r0.pkl', 'rb') as f:
    data_all = pickle.load(f)
```

```python
# check the loaded data before we run
print('Reward :', data_ghc[0]['R'])
print('Choice :', data_ghc[0]['choice'])
x = data_ghc[0]
```

```python
## lets fit all models to each group individually and to all data

# initalise data list an empty dicts for storing fits
data_list = [data_ghc, data_ghd, data_isoc, data_isod, data_all]
ghc, ghd, isoc, isod, all = {}, {}, {}, {}, {}

# make list of empty dicts for each model
frl_list = [ghc, ghd, isoc, isod, all]
cafrl_list = [ghc, ghd, isoc, isod, all]
safrl_list = [ghc, ghd, isoc, isod, all]
cafrl_A_list = [ghc, ghd, isoc, isod, all]
safrl_A_list = [ghc, ghd, isoc, isod, all]

# loop through each group and run frl
for i in range(len(data_list)):
    frl_list[i] = modelling.fit(data = data_list[i] ,nP = 2, fit_func = FRL.fit, n_jobs = 2,
                                fit_args = ['R','choice','S','dimension1'],transforms = ['sigmoid', 'exp'],fit = 'EM')
# loop through each group and run cafrl
for i in range(len(data_list)):
    cafrl_list[i] = modelling.fit(data = data_list[i] ,nP = 3, fit_func = CAFRL.fit, n_jobs = 2,
                                  fit_args = ['R','choice','S','dimension1'],transforms = ['sigmoid', 'exp', None],fit = 'EM')
# loop through each group and run safrl
for i in range(len(data_list)):
    safrl_list[i] = modelling.fit(data = data_list[i] ,nP = 4, fit_func = SAFRL.fit, n_jobs = 2,
                                  fit_args = ['R','choice','S','dimension1'],transforms = ['sigmoid','exp', None, 'sigmoid'],fit = 'EM')
# loop through each group and run cafrl
for i in range(len(data_list)):
    cafrl_A_list[i] = modelling.fit(data = data_list[i] ,nP = 3, fit_func = CAFRL_A.fit, n_jobs = 2,
                                  fit_args = ['R','choice','S','dimension1'],transforms = ['sigmoid', 'exp', 'sigmoid'],fit = 'EM')
# loop through each group and run safrl
for i in range(len(data_list)):
    safrl_A_list[i] = modelling.fit(data = data_list[i] ,nP = 4, fit_func = SAFRL_A.fit, n_jobs = 2,
                                  fit_args = ['R','choice','S','dimension1'],transforms = ['sigmoid','exp', 'sigmoid', 'sigmoid'],fit = 'EM')

# save fits 
groups = ['ghc', 'ghd', 'isoc', 'isod', 'all']

i=0
for fit in frl_list:
    with open('data/modelling/fits/frl/' + groups[i] +  '_r0.pkl', 'wb') as f:
        pickle.dump(fit, f)
    i+=1

i=0
for fit in cafrl_list:
    with open('data/modelling/fits/cafrl/' + groups[i] +  '_r0.pkl', 'wb') as f:
        pickle.dump(fit, f)
    i+=1

i=0
for fit in safrl_list:
    with open('data/modelling/fits/safrl/' + groups[i] +  '_r0.pkl', 'wb') as f:
        pickle.dump(fit, f)
    i+=1

i=0
for fit in cafrl_A_list:
    with open('data/modelling/fits/cafrl_A/' + groups[i] +  '_r0.pkl', 'wb') as f:
        pickle.dump(fit, f)
    i+=1

i=0
for fit in safrl_A_list:
    with open('data/modelling/fits/safrl_A/' + groups[i] +  '_r0.pkl', 'wb') as f:
        pickle.dump(fit, f)
    i+=1
```
